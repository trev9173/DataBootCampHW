```python
import pandas as pd
```

```python
hawaii_measurements = pd.read_csv('./hawaii_measurements.csv')
hawaii_stations = pd.read_csv('./hawaii_stations.csv')
```

```python
# Drop all rows with NA values
hawaii_stations
```

```python
# Drop all rows with NA values
dates = []
hawaii_measurements = hawaii_measurements.dropna()
hawaii_measurements = hawaii_measurements.reset_index()
hawaii_measurements = hawaii_measurements.drop(['index'], axis=1)
for row in hawaii_measurements['date']:
    new_date = pd.to_datetime(row)
    dates.append(new_date)
hawaii_measurements['date'] = dates
hawaii_measurements['date'] = hawaii_measurements['date'].dt.date
hawaii_measurements
```

```python
t = dates[1]
t.date()
```

```python
hawaii_measurements_means = hawaii_measurements.groupby(by='station').mean()
hawaii_measurements_means = hawaii_measurements_means.reset_index()
hawaii_measurements_means
```

```python
hawaii_combined = pd.merge(hawaii_measurements_means, hawaii_stations, how='inner', on='station')
```

```python
hawaii_combined.to_csv('./clean_hawaii_combined.csv')
hawaii_measurements.to_csv('./clean_hawaii_measurements.csv')
hawaii_stations.to_csv('./clean_hawaii_stations.csv')
```
